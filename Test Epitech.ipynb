```python
from sqlite3.dbapi2 import DatabaseError
from os import EX_DATAERR
import pandas as pd
#Initialisation persistence 
#creation base de donnÃ©e sqlite
import sqlite3
#Global Variables
con = sqlite3.connect('user.db')
model = {'name':'','company_name':'','email':'','phone':''}
Credential = {'login':'','pwd':''}
loged= False
currentUser =''
exit= False

def GeCnxDb() :
  return con.cursor()

def UpdateSchema():
  try:
    cur = GeCnxDb()
    cur.execute('''CREATE TABLE IF NOT EXISTS user
                (person_id INTEGER PRIMARY KEY, name text NOT NULL, company_name text, email text, phone real)''')
    cur.execute('''CREATE TABLE IF NOT EXISTS login
                (userName,pwd)''')
    cur.execute('''CREATE TABLE IF NOT EXISTS Cards
                (cardId INTEGER PRIMARY KEY,name text, company_name text, email text NOT NULL, phone real, userName text)''') #TODO ForeignKEy on table login 
  finally:
    con.commit()
    cur.close()
  
def IsUser(login, pwd, singleton):
  try:
    global currentUser
    request = "SELECT * FROM login where userName = ? and pwd = ?"
    singleton.execute(request,(login,pwd))
    data = singleton.fetchall() 
    if len(data) > 0 :
      currentUser = data[0][0]
      return True
    return False
  finally:
    con.commit()
    singleton.close()

def InsertUserInformation(singleton, user) :
  try:
    request = "INSERT INTO user VALUES (null,?,?,?,?)"
    singleton.execute(request,(user["name"],user["company_name"],user["email"],user["phone"]))
  finally:
    con.commit()
    singleton.close()

def InsertCredentialInformation(singleton, user) :
  try:
    request = "INSERT INTO login VALUES (?,?)"
    singleton.execute(request,(user["login"],user["pwd"]))
  finally:
    con.commit()
    singleton.close()

def InsertNewCard(singleton, model):
  try:
    global currentUser
    request = "INSERT INTO Cards VALUES (null,?,?,?,?,?)"
    singleton.execute(request,(model["name"],model["company_name"],model["email"],model["phone"],currentUser))
  finally:
    con.commit()
    singleton.close()
def PrintCardCurrentUser(singleton):
    global currentUser
    request ="SELECT name , company_name , email, phone FROM Cards WHERE userName =\'"+currentUser+"\'"
    df= pd.read_sql_query(request,con)
    print(df)

def AddBusinessCardsInterface():
  print('[ADD BUSINESS CARD] Enter the information about card')
  model['name'] = input("enter contact name")
  model['company_name'] = input("enter contact company name")
  while  model['email'] == '':
    model['email'] = input("Enter contact email (mandatory)")
  print('enter contact phone number')
  model['phone'] = str(input())
  InsertNewCard(GeCnxDb(),model)
def LogOnInterface() :
  login = input("Enter your login")
  pwd = input("Enter your password")
  if(IsUser(login,pwd,GeCnxDb())):
    global loged 
    loged = True
    print("Welcome "+login)
  else :
    print("Your User or Password are unknown")

def RegisterInterface() :
  print('Welcome to our plateform. Please enter your information')
  while model['name'] == '':
    model['name'] = input("enter your name (mandatory)")
    model['company_name'] = input("enter your company name (optional)")
    print('enter your email adress (optional)')
    model['email'] = input()
    print('enter your phone number (optional)')
    model['phone'] = str(input())
    while Credential['login'] =='' :
      print('Please enter a login (mandatory)')
      Credential['login']=str(input())
    while Credential['pwd'] =='' :
      print('Please enter a password (mandatory)')
      Credential['pwd']=str(input())
      InsertUserInformation(GeCnxDb(),model)
      InsertCredentialInformation(GeCnxDb(),Credential)   
#Initialisation BDD
UpdateSchema()
#Interface utilisateur

while not exit:
  if(loged):
    saisie = input("Press 1 to add business Card, Press 2 to see your business Card, Press 3 to Logout")
    if saisie == "1" :      
      AddBusinessCardsInterface() 
    elif saisie == "2" :
      print("[Cards Contact OF "+currentUser+"]")
      PrintCardCurrentUser(GeCnxDb())
    elif saisie == "3" :
      print('See u soon :)')
      loged =False
  else:
    saisie = input("Press 1 to login, 2 to register, 3 to exit")
    if saisie == "1" :
      LogOnInterface() 
    elif saisie == "2" :
      RegisterInterface()
    elif saisie == "3" :
      print('See u soon :)')
      exit =True
  

#User logged

```
